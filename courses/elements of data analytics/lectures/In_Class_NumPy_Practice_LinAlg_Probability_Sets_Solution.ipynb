## **Linear Algebra with NumPy**

#### **Matrix Creation and Operations**

Let's start by creating matrices and performing basic operations on them.

```python
# Creating matrices
A = np.array([[1,2,3], [4, 5, 6]])
B = np.array([[2,2,2], [2, 2,2]])
```

* **Add the two matrices we created**

```python
# Matrix addition
print("Matrix A + Matrix B =\n", A + B)
```

* **Let's do a matrix multiplication**

* How difficult can it be? 

```python
# Matrix multiplication
# print("\nA * B =\n", np.dot(A,B))
print("\nA * B =\n", np.dot(A.transpose(),B))
```

* You received a `matmul` ValueError complaining about a mismatch in our matrices' dimension. Here is why: 

* Matrix `A` has dimensions 2×3 and matrix `B` has dimensions 2×3. For matrix multiplication using the `np.dot` function (or the `@` operator), the number of columns in the first matrix must be equal to the number of rows in the second matrix. In this case, you can't multiply `A` and `B` directly since the number of columns in `A` (which is 3) does not match the number of rows in `B` (which is 2).

* However, you can multiply A with the transpose of B (denoted B<sup>T</sup> )
The transpose of a matrix switches its rows and columns. For matrix B above, 
B<sup>T</sup> will be a 3×2 matrix.

### Review of Matrix Transpose Operator

Given the matrices \( A \) and \( B \):


\begin{align*}
A =
\begin{bmatrix}
    1 & 2 & 3 \\
    4 & 5 & 6 \\
\end{bmatrix}
\end{align*}



\begin{align*}
B =
\begin{bmatrix}
    2 & 2 & 2 \\
    2 & 2 & 2 \\
\end{bmatrix}
\end{align*}


The transpose of \( B \), \( B<sup>T</sup>  \), is:

\( B<sup>T</sup> \):

\begin{align*}
B^T =
\begin{bmatrix}
    2 & 2 \\
    2 & 2 \\
    2 & 2 \\
\end{bmatrix}
\end{align*}


When you multiply \( A \) with \( B<sup>T</sup>  \) using `A @ B.T`, you're performing matrix multiplication between a \( 2 x 3 \) matrix and a \( 3 x 2 \) matrix. The resulting matrix will be \( 2 x 2 \).

On the other hand, directly multiplying \( A \) and \( B \) using `A @ B` would require \( B \) to have the same number of rows as \( A \) has columns.


$$\begin{bmatrix} 1 & 2 & 1 \\ 3 & 0 & 1 \\ 0 & 2 & 4 \end{bmatrix}$$

### **Correlation Matrix**


A correlation matrix is just a table with the correlation coefficients for different variables. The matrix shows how all the possible pairs of values in a table are related to each other. It is a powerful tool for summarizing a large data set and finding and showing patterns in the data.

* A correlation matrix provides the correlation coefficients between pairs of numerical features. Values close to 1 or -1 indicate a strong positive or negative correlation, respectively. A value close to 0 suggests little to no linear relationship between the features.

* Let's look at the numerical columns and investigate any correlations among them. 

```python
# Computing the correlation matrix using numpy's corrcoef method
# We'll transpose the data because corrcoef expects variables to be represented by rows, not columns.
# Returns the correlation coefficient matrix, where the diagonals are always 1 
correlation_matrix_np = np.corrcoef(df[numerical_columns].dropna().values, rowvar=False)


# You can also do it with Pandas' corr() method. This method returns a full correlation matrix.
#correlation_matrix = df[numerical_columns].corr()

correlation_matrix_np
```

* The result is a matrix where each entry at position (i,j) represents the correlation between the i<sup>th</sup>  and  j<sup>th</sup> feature. The diagonal elements are all 1 because any feature is perfectly correlated with itself.

**Please Note** 

* The `rowvar=False` argument in the `np.corrcoef()` function is responsible for treating columns as variables instead of rows. By setting rowvar=False, we're indicating that each column represents a variable, and each row is an observation of those variables. This is essentially equivalent to taking the transpose for the purpose of correlation calculation.

* In the traditional use of `np.corrcoef()`, rows are considered as variables and columns as observations. But in our dataset (and most tabular datasets), columns are the variables (features) and rows are the observations (samples). So, by setting `rowvar=False`, we're aligning the data correctly for correlation calculation without physically transposing the data.

### **Let's visualize this correlation matrix with a heatmap**


* The output of the correlation matrix using the numpy array above does not easily let us discern the correlations. Here, some visualization would be very helpful. 


A heatmap is a graphical representation that uses color shades to depict the strength and direction of correlations between variables. By presenting data across two dimensions, it provides an instant visual summary. It allows us to quickly discern patterns and the degree of relationship among data points.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the correlation matrix using seaborn's heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_np, annot=True, cmap='coolwarm', xticklabels=numerical_columns, yticklabels=numerical_columns)
plt.title("Correlation Matrix")
plt.show()
```

* The color scale on the right side indicates the strength and direction of the correlation: values close to 1 (darker red) indicate strong positive correlation, values close to -1 (darker blue) indicate strong negative correlation, and values close to 0 (lighter colors) indicate weak or no correlation.

* The labels on the `x` and `y` axes represent the numerical columns from the dataset.

* This visualization provides a quick way to identify relationships between pairs of variables in the dataset. Strongly correlated or anti-correlated variables are represented by distinctly colored squares.

### **Matrix Rank**

We learned that the rank of a matrix is the maximum number of linearly independent column vectors (or row vectors) in the matrix. 

```python
rank_matrix = np.linalg.matrix_rank(df[numerical_columns].dropna())
print(f'We have {rank_matrix} linearly independent rows or columns in this subset of the data.')
```

* **`np.linalg.matrix_rank(...)`** computes the matrix rank of the input. The matrix rank is a measure that tells us the number of linearly independent rows or columns in a matrix. In other words, it provides insight into the "information richness" of the matrix. A full-rank matrix is one where all rows (or columns) are linearly independent.

### **Probability**

Suppose automobiles are classified as "Wide Automobiles" if they have a width above the 75th percentile of all automobile widths in the dataset. What is the probability that a randomly selected automobile from the dataset is a "Wide Automobile"?

To solve this using NumPy:

* Calculate the 75th percentile for the "Avg. Area Number of Rooms" column.
* Count the number of houses with an "Avg. Area Number of Rooms" above this value.
* Divide this count by the total number of houses to get the probability.

```python
percentile_75_width = np.percentile(df['width'].dropna(), 75)

# Count the number of automobiles with "width" above this value
wide_automobile_count = np.sum(df['width'] > percentile_75_width)

# Compute the probability
probability_wide_automobile = wide_automobile_count / len(df.dropna(subset=['width']))

percentile_75_width, wide_automobile_count, probability_wide_automobile
```
