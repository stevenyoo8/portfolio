### 2. **Removing Duplicates**

```python
initial_count = len(modified_df)
modified_df = modified_df.drop_duplicates()
remaining_count = len(modified_df)

initial_count, remaining_count
```

### 3. **Correcting Inconsistent Values**

```python
inconsistent_values = modified_df['fuel-type'].value_counts()

modified_df['fuel-type'] = modified_df['fuel-type'].replace('water', 'gas')

inconsistent_values, modified_df['fuel-type'].value_counts()
```

### 4. **Handling Outliers**

Interquartile Range (IQR) can be used to detect outliers. A common rule of thumb is that any data point that falls below Q1 - 1.5 x IQR or above Q3 + 1.5 X IQR might be considedred an outlier.

```python
# Use IQR to identify and remove outliers in the 'price' column
Q1 = modified_df['price'].quantile(0.25)
Q3 = modified_df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
modified_df_no_outliers = modified_df[(modified_df['price'] >= lower_bound) & (modified_df['price'] <= upper_bound)]

len(modified_df), len(modified_df_no_outliers)
```
