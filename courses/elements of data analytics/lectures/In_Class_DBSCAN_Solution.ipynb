### **DBSCAN in Python**

We will use the DBSCAN implementation from the `sklearn` library.

* Let's start by generating some synthetic data and then apply DBSCAN to it.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate synthetic data
X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)

# Apply DBSCAN
db = DBSCAN(eps=0.3, min_samples=5)
labels = db.fit_predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering')
plt.show()
```

### **Exercise 1**

* Using the above code as a reference, apply DBSCAN on a dataset with blobs of points. Experiment with different values of eps and min_samples and observe the results.

```python
from sklearn.datasets import make_blobs

# Generate synthetic data with blobs
X, _ = make_blobs(n_samples=300, centers=3, random_state=0, cluster_std=0.6)

# Apply DBSCAN
db = DBSCAN(eps=0.5, min_samples=5)
labels = db.fit_predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering with Blobs')
plt.show()
```

### **Exercise 2**

* Using the same dataset from Exercise 1, compare the results of DBSCAN with another clustering algorithm, like KMeans. What are the key differences you notice?

```python
from sklearn.cluster import KMeans

# Generate synthetic data
X, _ = make_moons(n_samples=300, noise=0.05, random_state=0)

# Apply KMeans
kmeans = KMeans(n_clusters=2)
kmeans_labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title('KMeans Clustering with Moons')
plt.show()

# Apply DBSCAN
db = DBSCAN(eps=0.3, min_samples=5)
labels = db.fit_predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering with Moons')
plt.show()

# Observations:
# KMeans struggles with the crescent moon shapes and tries to cut through them.
# DBSCAN, on the other hand, correctly identifies the two separate moon shapes.
```

### **Exercise 3**

* Run DBSCAN on the dataset but add some random noise (data points) outside the clusters. How does the noise impact the DBSCAN results? How might adjusting the parameters help?

```python
# Add noise to the dataset
noise = np.random.rand(20, 2) * 10 - 5  # 20 random points between [-5, 5]
X_with_noise = np.vstack([X, noise])

# Apply DBSCAN
db = DBSCAN(eps=0.5, min_samples=5)
labels = db.fit_predict(X_with_noise)

# Plot the results
plt.scatter(X_with_noise[:, 0], X_with_noise[:, 1], c=labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering with Noise')
plt.show()

# Observations:
# DBSCAN will usually label random noise as '-1', making it easy to identify and filter out.
# Adjusting the eps and min_samples can help in correctly classifying noisy data, but it's a balance.
```
