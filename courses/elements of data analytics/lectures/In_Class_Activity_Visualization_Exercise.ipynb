```python
# Import pandas and matplotlib (please note, you will need to install matplotlib to your machine if you haven't already)
import pandas as pd
import matplotlib.pyplot as plt #pyploy is one of the components of matplotlib
```

```python
# Create a dataframe from our dataset
df = pd.read_csv('flights.csv', index_col=False)
```

```python
# Let's remember the column names
df.columns
```

```python
# Create a histogram of departure delays, titled "Distribution of Departure Delays"

# Define the size of the plot
plt.figure(figsize=(10, 6))

# Get the data from the dataframe (and remove NaNs) and choose histogram (`hist`) as your plot
plt.hist(df['dep_delay'].dropna(), bins=50, color='blue', edgecolor='black')

plt.title('Distribution of Departure Delays')
plt.xlabel('Delay (minutes)')
plt.ylabel('Number of Flights')

# Add horizontal lines to the plot, aligned with the y-axis tick marks, to make the plot easier to read
plt.grid(True, axis='y')

# Save the plot to your machine (do this before plt.show())
# plt.savefig('departure_delays.png')


# Display the plot
plt.show()
```

```python
# Let's create a histogram using pandas
df['dep_delay'].hist()
```

```python
# Create a bar chart titled Average Departure Delay by Airline Carrier
avg_delay_by_carrier = df.groupby('carrier')['dep_delay'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
print(avg_delay_by_carrier)

# retrieve the airline carriers, which are used as the x-coordinates, with the "indices" of the Series and
# retrieve the average delays, the actual "values" of the Series, which will constitute the y-coordinates
plt.bar(avg_delay_by_carrier.index, avg_delay_by_carrier.values, color='green')

#TODO: Complete the plot
plt.title('Avg Departure Delay by Carrier')
plt.xlabel('Carrier')
plt.ylabel('Departure Delay (minutes)')

plt.grid(True, axis = 'y')
```

```python
# Create a scatter plot of air time vs. distance, titled "Air Time vs Distance"
plt.figure(figsize=(10, 6))
plt.scatter(df['air_time'], df['distance'], alpha=0.5, color='red')

#TODO: Complete the plot
plt.title('Air Time vs Distance')
plt.xlabel('Air Time (min)')
plt.ylabel('Distance (mi)')

plt.grid(True)
```

```python
# Create a bar chart of the number of flights from each origin, titled "Number of Flights From Each Origin"
plt.figure(figsize=(10, 6))
flights_by_origin = df['origin'].value_counts()

# TODO: Complete the plot
# retrieve the flights by origin, which are used as the x-coordinates, with the "indices" of the Series and
# retrieve the actual "values" of the Series, which will constitute the y-coordinates
plt.bar(flights_by_origin.index, flights_by_origin.values, color='green')

#TODO: Complete the plot
plt.title('Number of Flights From Each Origin')
plt.xlabel('Origin')
plt.ylabel('Number of Flights')
```

### **Review at Home**

More awesome stuff you can do with Matplotlib

* Another great way of creating plots (especially for research papers, business proposals, etc. where you might want to display multiple visualizations on your data) is using `subplots` from `pyplot` module in Matplotlib.

* `plt.subplots()`: This function is a part of Matplotlib's `pyplot` module and is used to create a figure and a grid of subplots with a single call.

```python
# Set up the figure and axis for subplots
fig, axs = plt.subplots( 2, 2, figsize=(15, 10)) # a 2 x 2 subplot
fig.tight_layout(pad=5.0) # tight_layout ensures that there is enough space between subplots

# Create a histogram of departure delays to see the distribution of delays.
axs[0, 0].hist(df['dep_delay'].dropna(), bins=50, color='blue', edgecolor='black')
axs[0, 0].set_title('Distribution of Departure Delays')
axs[0, 0].set_xlabel('Delay (minutes)')
axs[0, 0].set_ylabel('Number of Flights')
axs[0, 0].grid(True, axis='y')

# Create a bar chart of average delay by airline carrier
avg_delay_by_carrier = df.groupby('carrier')['dep_delay'].mean().sort_values(ascending=False)
axs[0, 1].bar(avg_delay_by_carrier.index, avg_delay_by_carrier.values, color='green')
axs[0, 1].set_title('Average Departure Delay by Airline Carrier')
axs[0, 1].set_xlabel('Carrier')
axs[0, 1].set_ylabel('Average Delay (minutes)')
axs[0, 1].grid(True, axis='y')

# Create a scatter plot of air time vs. distance
axs[1, 0].scatter(df['air_time'], df['distance'], alpha=0.5, color='red')
axs[1, 0].set_title('Air Time vs. Distance')
axs[1, 0].set_xlabel('Air Time (minutes)')
axs[1, 0].set_ylabel('Distance (miles)')
axs[1, 0].grid(True)

# Create a bar chart of the number of flights from each origin
flights_by_origin = df['origin'].value_counts()
axs[1, 1].bar(flights_by_origin.index, flights_by_origin.values, color='purple')
axs[1, 1].set_title('Number of Flights From Each Origin')
axs[1, 1].set_xlabel('Origin')
axs[1, 1].set_ylabel('Number of Flights')
axs[1, 1].grid(True, axis='y')

# Display the visualizations
plt.show()
```
