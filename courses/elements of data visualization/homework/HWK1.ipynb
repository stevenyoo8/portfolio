**Q. 2 (10 points)**  
For each year find the maximum number of immigrants and the country that had the maximum number of immigrants. Create a new dataframe having three columns - year, country, and number of immigrants. For a given year, if there are two or more countries having the same maximum number of immigrants, then choose that country that comes first in alphabetical order.

```python
max_df = pd.DataFrame(columns=['Year', 'Country', 'Number of Immigrants'])

for year in range(1980, 2014):
    max_country = df.iloc[:-1,][str(year)].idxmax()
    max_immigrants = df.iloc[:-1,][str(year)].max()

    new_row = {'Year': year, 'Country': max_country, 'Number of Immigrants': max_immigrants}
    max_df = max_df.append(new_row, ignore_index=True)

max_df.head()
```

**Q. 3 (10 points)**   
For each year find the total, mean, standard deviation, maximum, minimum, and range (max - min). Create a new dataframe having columns - year, total, mean, standard deviation, maximum, minimum, and range.

```python
summary_df = pd.DataFrame(columns=['Year', 'Total', 'Mean', 'Standard Deviation', 'Maximum', 'Minimum', 'Range'])

for year in range(1980, 2014):
    total_imm = df[str(year)].max()
    mean_imm = df.iloc[:-1,][str(year)].mean()
    std_imm = df.iloc[:-1,][str(year)].std()
    max_imm = df.iloc[:-1,][str(year)].max()
    min_imm = df.iloc[:-1,][str(year)].min()
    range_imm = max_imm - min_imm

    summary_df = summary_df.append({'Year': year, 'Total': total_imm, 'Mean': mean_imm, 'Standard Deviation': std_imm, 'Maximum': max_imm, 
                                  'Minimum': min_imm, 'Range': range_imm}, ignore_index=True)

summary_df.head()
```

**Q. 6 (10 points)**.  
Which country in each of the continents - Africa, Asia, Europe, Latin America, and Oceania had the largest total immigration? Print the continent, country, and total immigration.

```python
continent_list = df['Continent'].dropna().unique().tolist()

for continent in sorted(continent_list):
    new_df = df[df['Continent'] == str(continent)]
    max_immigration_country = new_df.iloc[:]['Total'].idxmax()
    total_immigration = new_df.loc[str(max_immigration_country)]['Total']
    print("Continent: " + str(continent))
    print("Country: " + str(max_immigration_country))
    print("Total Immigration: " + str(total_immigration))
    print()
```

**Q. 8 (10 points)**    
For the countries in the Region named as Caribbean find the total immigration in the two decades (1980 to 1989) and (1990 - 1999). Print the decades as (Eighties and Nineties) and the total immigration for these decades.

```python
new_df = df[df['Region'] == 'Caribbean']
new_df.loc['Total_Immigration'] = new_df.loc[:, '1980':'2013'].sum(numeric_only=True)

decade1_immigration = 0
for year in range(1980, 1990):
    decade1_immigration += new_df.loc['Total_Immigration'][str(year)].sum()

decade2_immigration = 0
for year in range(1990, 2000):
    decade2_immigration += new_df.loc['Total_Immigration'][str(year)].sum()

print("Eighties")
print("Total Immigration: " + str(decade1_immigration))
print()
print("Nineties")
print("Total Immigration: " + str(decade2_immigration))
```

**Q. 9 (10 points)**.  
The Scandinavian countries are Denmark, Norway, and Sweden. For each of the years (from 1980 to 2013) find the difference between the country with the highest immigration and the country with the lowest immigration. Print the year and the difference.

```python
new_df = df[(df.index == 'Denmark') | (df.index == 'Norway') | (df.index == 'Sweden')]

for year in range(1980, 2014):
    max_country = new_df[str(year)].idxmax()
    min_country = new_df[str(year)].idxmin()
    max_immigration = new_df[str(year)].max()
    min_immigration = new_df[str(year)].min()
    difference = max_immigration - min_immigration
    print("Year: " + str(year))
    print("Difference: " + str(difference))
    print()
```
