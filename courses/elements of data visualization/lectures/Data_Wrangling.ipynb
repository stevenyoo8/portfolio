### Data Wrangling

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the data file
df = pd.read_csv('cars.csv')
```

```python
# get the size of the dataframe
df.shape
```

```python
# get the head
df.head(5)
```

```python
# get the tail of the dataframe
df.tail(5)
```

```python
# replace '?' with NaN
df.replace('?', np.nan, inplace = True)
```

```python
# check if the replacements have gone through
df.head(10)
```

##### Detect Missing Data

```python
# find missing data
missing_data = df.isnull()
missing_data.head(10)
```

```python
# size of data frame missing_data
missing_data.shape
```

```python
# count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print(" ")
```

```python
# compute the mean for normalized-losses
avg_norm_loss = df['normalized-losses'].astype('float').mean()
print('Average of normalized-losses', avg_norm_loss)
```

```python
# replace missing normalized-losses with mean
df['normalized-losses'].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
# find the frequency of values for num-of-doors
df['num-of-doors'].value_counts()
```

```python
# replace missing values of 'num-of-doors' with the most frequent (4)
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
# check if the replacements have gone through
df['num-of-doors'].value_counts()
```

```python
# drop rows that do not have the price of cars
df.dropna(subset = ['price'], axis = 0, inplace = True)

# reset indices because we dropped/removed rows
df.reset_index(drop = True, inplace = True)
```

##### Check Correct Format

```python
# check data types
df.dtypes
```

```python
# convert data types to correct format
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
```
