```python
# Which country has donated the most amount of money (usd)?
donation_by_country = df.groupby('donor')['commitment_amount_usd_constant'].sum().sort_values(ascending = False)
plt.figure(figsize= (12,6))
plt.bar(donation_by_country.index, donation_by_country.values)
plt.xticks(rotation = 60)
plt.ticklabel_format(style='plain', axis='y')

plt.title('Donation Amount By Country')
plt.xlabel('Country')
plt.ylabel('Donation Amount (USD)')

plt.grid(True, axis = 'y')

plt.show()
```

```python
# What purposes are the most of the donation money being used for?
df_US = df[df['recipient'] == "United States"]
top_5_donation_by_purpose = df.groupby('coalesced_purpose_name')['commitment_amount_usd_constant'].sum().sort_values(ascending = False)[:5,]

plt.figure(figsize= (12,6))
plt.bar(top_5_donation_by_purpose.index, top_5_donation_by_purpose.values)
plt.xticks(rotation = 45) 
plt.ticklabel_format(style='plain', axis='y')

plt.title('Donation Amount by Purpose')
plt.xlabel('Purpose')
plt.ylabel('Donation Amount (USD)')

plt.grid(True, axis = 'y')

plt.show()
```
